```python
import pyspedas
import pytplot
from pytplot import tplot, options, store_data, data_quants

import numpy as np
import plasma_params as pp
```

```python
tr = ['2017-04-24', '2017-04-25']
from pyspedas.erg import mgf, pwe_ofa
mgf(trange=tr)
pwe_ofa(trange=tr, datatype='spec')
```

```python
q = pp.Q
me = pp.ME
B = data_quants['erg_mgf_l2_magt_8sec']
```

```python
fce = q * B * 1e-9 / (2 * np.pi * me) * 1e-3 #[kHz]
store_data('fce', data={'x': B['time'], 'y': fce})
```

```python
mh = pp.MH
fch = q * B * 1e-9 * 1e-3 / (2 * np.pi * mh) #[kHz]
store_data('fch', data={'x': B['time'], 'y': fch})
```

```python
from pyspedas.erg import orb
orb(trange=tr, level='l3')
```

```python
pytplot.tplot_names()
```

```python
store_data('fce_05', data={'x': B['time'], 'y': fce/2})
store_data('fce_01', data={'x': B['time'], 'y': fce/10})
store_data('fLHR', data={'x': B['time'], 'y': np.sqrt(fce) * np.sqrt(fch)})
```

```python
options('fce', opt_dict = {'ylog':1, 'line_style':':', 'thick':2, 'color':'black'})
options('fce_05', opt_dict = {'ylog':1, 'line_style':':', 'thick':2, 'color':'red'})
options('fce_01', opt_dict = {'ylog':1, 'line_style':':', 'thick':2, 'color':'magenta'})
options('fLHR', opt_dict = {'ylog':1, 'line_style':':', 'thick':2, 'color':'darkgreen'})
```

```python
options('erg_pwe_ofa_l2_spec_E_spectra_132', opt_dict={'yrange':[3e-2, 2e1]})
store_data('ofaE_f', data = ['erg_pwe_ofa_l2_spec_E_spectra_132', 'fce', 'fce_05', 'fce_01', 'fLHR', 'erg_pwe_ofa_l2_spec_E_spectra_132'])
```

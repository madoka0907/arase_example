```python
tr = ['2017-04-24', '2017-04-25']
from pyspedas.erg import mgf, pwe_hfa, pwe_ofa, pwe_efd
mgf(trange=tr)
pwe_hfa(trange=tr, datatype='spec')
pwe_ofa(trange=tr, datatype='spec')
pwe_efd(trange=tr, datatype='spec')
```

$\omega_c = \frac{qB}{m_e}$\
$f_c = \frac{qB}{2\pi  m_e}$

q:電荷量
B:磁場
m_e:電子の質量

```python
q = pp.Q
me = pp.ME
B = data_quants['erg_mgf_l2_magt_8sec']
```

```python
B
```

data_quantsで変数をxarrayの形にできる\
軸に名前がある行列にできる\


```python
np.pi
```

T\
'erg_mgf_l2_magt_8sec'にはnTの単位でデータが入っているので実際の磁場の大きさに直すには10のマイナス9乗をかける必要がある

```python
fce = q * B * 1e-9 / (2 * np.pi * me) * 1e-3 #[kHz]
```

```python
store_data('fce', data={'x': B['time'], 'y': fce})
```

```python
mh = pp.MH
fch = q * B * 1e-9 / (2 * np.pi * mh) #[Hz]
```

```python
mhe = pp.MHE
fche = q * B * 1e-9 / (2 * np.pi * mhe)
```

```python
store_data('fch', data={'x': B['time'], 'y': fch})
```

```python
store_data('fche', data={'x': B['time'], 'y': fche})
```

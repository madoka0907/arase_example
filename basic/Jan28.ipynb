```python
import pyspedas
import pytplot
from pytplot import tplot, options, store_data, data_quants, xlim

import numpy as np
import plasma_params as pp
```

```python
pp.Q
np.pi
```

```python
tr = ['2017-04-24 14:00:00', '2017-04-24 17:00:00']
from pyspedas.erg import mgf, pwe_hfa, pwe_ofa, orb
orb(trange=tr)
mgf(trange=tr)
pwe_hfa(trange=tr, level='l3')
pwe_hfa(trange=tr, datatype='spec')
pwe_ofa(trange=tr, datatype='spec')
```

```python
q = pp.Q
me = pp.ME
B = data_quants['erg_mgf_l2_magt_8sec']
```

```python
fce = q * B * 1e-9 / (2 * np.pi * me) * 1e-3 #[kHz]
store_data('fce', data={'x': B['time'], 'y': fce})
options('fce', opt_dict = {'ylog':1, 'line_style':':', 'thick':2, 'color':'black'})
store_data('fce_01', data={'x': B['time'], 'y': fce/10})
options('fce_01', opt_dict = {'ylog':1, 'line_style':':', 'thick':2, 'color':'magenta'})
```
